# Mission 1

Display the following information [for this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user).
- The last 5 rows
- The first 5 rows
- The type of each column

Note : start coding something by typing `pd.read_csv()` and explore the different options available, especially the `sep` parameter.


```
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user', sep = "|")

print("Les 5 dernières lignes :")
print(df.iloc[-5:])

print("\nLes 5 premières lignes :")
print(df.iloc[:5])

print("\nType de chaque colonne :")
print(df.dtypes)
```
