**Partie 1 : G√©n√©ration de Contenu avec OpenAI**

1. Configurer LangChain avec OpenAI :
- Installez les biblioth√®ques n√©cessaires (langchain, langchain-openai, python-dotenv).
- Obtenez votre cl√© API OpenAI et configurez LangChain pour l'utiliser.

```python
import os
from dotenv import load_dotenv
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
```

```python
load_dotenv()
openai_key = os.getenv('OPENAI_API_KEY')
langchain_key = os.getenv('LANGCHAIN_API_KEY')
```

2. T√¢che 1 : G√©n√©ration d'un Sc√©nario pour un Jeu Vid√©o :
- √âcrivez un script qui g√©n√®re un sc√©nario cr√©atif pour un jeu vid√©o de science-fiction, comprenant une description du monde, des personnages principaux, et une intrigue de base.
- Testez diff√©rentes valeurs pour les param√®tres temperature (par exemple, de 0.3 √† 1.0) et max_tokens pour explorer comment ces ajustements influencent la cr√©ativit√© et la longueur du sc√©nario.


```python
# Cr√©ation du mod√®le avec temperature et max_tokens par d√©faut
llm = ChatOpenAI(
    openai_api_key=openai_key,
    model_name="gpt-4o-mini",
    temperature=0.7,
    max_tokens=500
)

# D√©finition du prompt
prompt = PromptTemplate(
    input_variables=["genre"],
    template="""
Tu es un sc√©nariste expert en jeux vid√©o. √âcris un sc√©nario cr√©atif pour un jeu vid√©o de type {genre}, avec :
- Une description du monde futuriste
- Les personnages principaux
- Une intrigue captivante

Sc√©nario :
"""
)

# Cr√©ation de la cha√Æne LangChain
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
genres = ["science-fiction", "fantasy", "post-apocalyptique"]
temperatures = [0.3, 0.7, 1.0]
tokens = [300, 600]

for temp in temperatures:
    for max_tok in tokens:
        print(f"\n--- Temp√©rature {temp} | Max tokens {max_tok} ---")
        llm = ChatOpenAI(openai_api_key=openai_key, temperature=temp, max_tokens=max_tok)
        chain = LLMChain(llm=llm, prompt=prompt)
        result = chain.run("science-fiction")
        print(result)
```

2. T√¢che 2 : G√©n√©ration de R√©ponses en Temps R√©el :
- Utilisez le mode de streaming pour afficher les r√©ponses en temps r√©el. Simulez une interaction en direct o√π l'assistant r√©agit aux entr√©es des utilisateurs.

```python
streaming_llm = ChatOpenAI(
    openai_api_key=openai_key,
    streaming=True,
    temperature=0.7,
    callbacks=[StreamingStdOutCallbackHandler()],
    max_tokens=200
)

memory = ConversationBufferMemory()
conversation = ConversationChain(
    llm=streaming_llm,
    memory=memory,
    verbose=True
)

conversation.predict(input="Bonjour, peux-tu me raconter une anecdote dr√¥le ?")
```

**Partie 2 : Traduction et Personnalisation avec Mistral**

1. Configurer LangChain avec Mistral :
- Installez les biblioth√®ques n√©cessaires (langchain_mistralai).
- Obtenez votre cl√© API Mistral et configurez LangChain pour l'utiliser.

```python
from langchain_mistralai import ChatMistralAI
mistral_key = os.getenv('MISTRAL_API_KEY')
```

2. T√¢che 1 : Traduction Cr√©ative :
- Utilisez Mistral pour traduire des textes po√©tiques ou litt√©raires de l'anglais au fran√ßais. Manipulez le param√®tre temperature pour ajuster le niveau de cr√©ativit√© de la traduction, et donnez des consignes sp√©cifiques pour influencer le style (par exemple, "traduisez de mani√®re po√©tique").

```python
# Liste des phrases √† traduire
phrases = [
    "The stars whispered secrets to the lonely night.",
    "She wandered through dreams, chasing forgotten melodies."
]

# Diff√©rentes temp√©ratures √† tester
temperatures = [0.3, 0.7, 1.0]

# üîÅ Boucle sur chaque temp√©rature et phrase
for temp in temperatures:
    print(f"\n=== Temp√©rature {temp} ===")
    llm = ChatMistralAI(
        model="mistral-large-latest",
        temperature=temp,
        max_retries=2,
    )

    for phrase in phrases:
        messages = [
            ("system", "Vous √™tes un traducteur po√©tique. Traduisez le texte suivant de mani√®re lyrique de l'anglais au fran√ßais."),
            ("human", phrase)
        ]

        response = llm.invoke(messages)
        print(f"\nüßæ Texte original : {phrase}")
        print(f"üé® Traduction po√©tique : {response.content}")
```

3. T√¢che 2 : Conversation Contextuelle :
- Cr√©ez un assistant de chat qui se souvient du contexte de la conversation. - Testez diff√©rentes valeurs pour max_retries et top_p pour am√©liorer la coh√©rence et la diversit√© des r√©ponses en fonction des messages pr√©c√©dents.

```python
# Param√®tres √† ajuster
top_p = 0.9
max_retries = 3

# Historique simul√© (contexte)
conversation = [
    ("system", "Tu es un assistant litt√©raire qui discute de livres avec l'utilisateur."),
    ("human", "Bonjour, peux-tu me recommander un roman captivant ?"),
    ("ai", "Bien s√ªr ! Je te recommande *L'√âl√©gance du h√©risson* de Muriel Barbery."),
    ("human", "Pourquoi ce livre en particulier ?")
]

# Cr√©ation du mod√®le avec param√®tres
llm = ChatMistralAI(
    model="mistral-large-latest",
    temperature=0.7,
    top_p=top_p,
    max_retries=max_retries,
)

# Appel avec tout le contexte
response = llm.invoke(conversation)

print("ü§ñ R√©ponse avec contexte :")
print(response.content)
```
