**Partie 1 : Génération de Contenu avec OpenAI**

1. Configurer LangChain avec OpenAI :
- Installez les bibliothèques nécessaires (langchain, langchain-openai, python-dotenv).
- Obtenez votre clé API OpenAI et configurez LangChain pour l'utiliser.

```python
import os
from dotenv import load_dotenv
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain_openai import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
```

```python
load_dotenv()
openai_key = os.getenv('OPENAI_API_KEY')
langchain_key = os.getenv('LANGCHAIN_API_KEY')
```

2. Tâche 1 : Génération d'un Scénario pour un Jeu Vidéo :
- Écrivez un script qui génère un scénario créatif pour un jeu vidéo de science-fiction, comprenant une description du monde, des personnages principaux, et une intrigue de base.
- Testez différentes valeurs pour les paramètres temperature (par exemple, de 0.3 à 1.0) et max_tokens pour explorer comment ces ajustements influencent la créativité et la longueur du scénario.


```python
# Création du modèle avec temperature et max_tokens par défaut
llm = ChatOpenAI(
    openai_api_key=openai_key,
    model_name="gpt-4o-mini",
    temperature=0.7,
    max_tokens=500
)

# Définition du prompt
prompt = PromptTemplate(
    input_variables=["genre"],
    template="""
Tu es un scénariste expert en jeux vidéo. Écris un scénario créatif pour un jeu vidéo de type {genre}, avec :
- Une description du monde futuriste
- Les personnages principaux
- Une intrigue captivante

Scénario :
"""
)

# Création de la chaîne LangChain
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
genres = ["science-fiction", "fantasy", "post-apocalyptique"]
temperatures = [0.3, 0.7, 1.0]
tokens = [300, 600]

for temp in temperatures:
    for max_tok in tokens:
        print(f"\n--- Température {temp} | Max tokens {max_tok} ---")
        llm = ChatOpenAI(openai_api_key=openai_key, temperature=temp, max_tokens=max_tok)
        chain = LLMChain(llm=llm, prompt=prompt)
        result = chain.run("science-fiction")
        print(result)
```

2. Tâche 2 : Génération de Réponses en Temps Réel :
- Utilisez le mode de streaming pour afficher les réponses en temps réel. Simulez une interaction en direct où l'assistant réagit aux entrées des utilisateurs.

```python
streaming_llm = ChatOpenAI(
    openai_api_key=openai_key,
    streaming=True,
    temperature=0.7,
    callbacks=[StreamingStdOutCallbackHandler()],
    max_tokens=200
)

memory = ConversationBufferMemory()
conversation = ConversationChain(
    llm=streaming_llm,
    memory=memory,
    verbose=True
)

conversation.predict(input="Bonjour, peux-tu me raconter une anecdote drôle ?")
```

**Partie 2 : Traduction et Personnalisation avec Mistral**

1. Configurer LangChain avec Mistral :
- Installez les bibliothèques nécessaires (langchain_mistralai).
- Obtenez votre clé API Mistral et configurez LangChain pour l'utiliser.

```python
from langchain_mistralai import ChatMistralAI
mistral_key = os.getenv('MISTRAL_API_KEY')
```

2. Tâche 1 : Traduction Créative :
- Utilisez Mistral pour traduire des textes poétiques ou littéraires de l'anglais au français. Manipulez le paramètre temperature pour ajuster le niveau de créativité de la traduction, et donnez des consignes spécifiques pour influencer le style (par exemple, "traduisez de manière poétique").

```python
# Liste des phrases à traduire
phrases = [
    "The stars whispered secrets to the lonely night.",
    "She wandered through dreams, chasing forgotten melodies."
]

# Différentes températures à tester
temperatures = [0.3, 0.7, 1.0]

# 🔁 Boucle sur chaque température et phrase
for temp in temperatures:
    print(f"\n=== Température {temp} ===")
    llm = ChatMistralAI(
        model="mistral-large-latest",
        temperature=temp,
        max_retries=2,
    )

    for phrase in phrases:
        messages = [
            ("system", "Vous êtes un traducteur poétique. Traduisez le texte suivant de manière lyrique de l'anglais au français."),
            ("human", phrase)
        ]

        response = llm.invoke(messages)
        print(f"\n🧾 Texte original : {phrase}")
        print(f"🎨 Traduction poétique : {response.content}")
```

3. Tâche 2 : Conversation Contextuelle :
- Créez un assistant de chat qui se souvient du contexte de la conversation. - Testez différentes valeurs pour max_retries et top_p pour améliorer la cohérence et la diversité des réponses en fonction des messages précédents.

```python
# Paramètres à ajuster
top_p = 0.9
max_retries = 3

# Historique simulé (contexte)
conversation = [
    ("system", "Tu es un assistant littéraire qui discute de livres avec l'utilisateur."),
    ("human", "Bonjour, peux-tu me recommander un roman captivant ?"),
    ("ai", "Bien sûr ! Je te recommande *L'Élégance du hérisson* de Muriel Barbery."),
    ("human", "Pourquoi ce livre en particulier ?")
]

# Création du modèle avec paramètres
llm = ChatMistralAI(
    model="mistral-large-latest",
    temperature=0.7,
    top_p=top_p,
    max_retries=max_retries,
)

# Appel avec tout le contexte
response = llm.invoke(conversation)

print("🤖 Réponse avec contexte :")
print(response.content)
```
