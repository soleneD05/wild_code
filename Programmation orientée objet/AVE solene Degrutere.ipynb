Consigne : Dans le notebook, crée une fonction qui prend en paramètre un message sous forme de liste ou de chaine de caractère, et affiche le message correctement (décrypté). Voici le message à décrypter : mlspjpahapvuz, jlzhy zlyhpa mply kl avp, zp zlbsltlua ps jvuuhpzzhpa wfaovu.

```python
def decrypter_message(message):
    # Vérifie si le message est une liste et le transforme en chaîne
    if isinstance(message, list):
        message = ''.join(message)

    decalage = -7
    # Décrypter le message
    decrypted_message = ''
    for char in message:
        if char.isalpha():  # Traiter uniquement les lettres
            decrypted_char = chr((ord(char) - ord('a') + decalage) % 26 + ord('a'))
            decrypted_message += decrypted_char
        else:
            # Garder les espaces
            decrypted_message += char

    # Afficher le message décrypté
    print("Message décrypté :", decrypted_message)

# Message à décrypter
message = "mlspjpahapvuz, jlzhy zlyhpa mply kl avp, zp zlbsltlua ps jvuuhpzzhpa wfaovu."

# Appeler la fonction pour décrypter et afficher le message
decrypter_message(message)
```
