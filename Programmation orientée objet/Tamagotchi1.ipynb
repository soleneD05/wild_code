```python
# Variables que nous voulons suivre pour le chien :
# 1. Nom
# 2. Énergie
# 3. Faim
# 4. Bonheur
# 5. Propreté

# Activités souhaitées et leurs impacts :
# 1. Jouer : Augmente le bonheur, diminue l'énergie.
# 2. Manger : Diminue la faim, augmente la propreté.
# 3. Dormir : Augmente l'énergie, diminue la faim.
# 4. Se promener : Augmente le bonheur, diminue l'énergie, et diminue légèrement la propreté.
# 5. Prendre un bain : Augmente la propreté, diminue légèrement l'énergie.
```

```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.energy = 100  # Énergie maximale : 100
        self.hunger = 0    # Faim minimale : 0
        self.happiness = 100  # Bonheur maximal : 100
        self.cleanliness = 100  # Propreté maximale : 100

    def play(self):
        if self.energy >= 20:
            self.happiness = min(self.happiness + 20, 100)
            self.energy = max(self.energy - 20, 0)
            print(f"{self.name} a joué et s'amuse beaucoup!")
        else:
            print(f"{self.name} est trop fatigué(e) pour jouer!")

    def eat(self):
        if self.hunger > 0:
            self.hunger = max(self.hunger - 30, 0)
            self.cleanliness = max(self.cleanliness - 10, 0)
            print(f"{self.name} a mangé et se sent rassasié(e).")
        else:
            print(f"{self.name} n'a pas faim en ce moment.")

    def sleep(self):
        if self.energy < 100:
            self.energy = min(self.energy + 50, 100)
            self.hunger = min(self.hunger + 20, 100)
            print(f"{self.name} a bien dormi et se sent reposé(e).")
        else:
            print(f"{self.name} n'est pas fatigué(e) pour dormir.")

    def walk(self):
        if self.energy >= 15:
            self.happiness = min(self.happiness + 15, 100)
            self.energy = max(self.energy - 15, 0)
            self.cleanliness = max(self.cleanliness - 10, 0)
            print(f"{self.name} a fait une belle promenade et a adoré ça!")
        else:
            print(f"{self.name} est trop fatigué(e) pour se promener!")

    def take_bath(self):
        if self.cleanliness < 100:
            self.cleanliness = min(self.cleanliness + 40, 100)
            self.energy = max(self.energy - 10, 0)
            print(f"{self.name} est maintenant tout(e) propre!")
        else:
            print(f"{self.name} est déjà propre!")

    def status(self):
        print(f"\nStatut de {self.name} :")
        print(f"\tÉnergie : {self.energy}")
        print(f"\tFaim : {self.hunger}")
        print(f"\tBonheur : {self.happiness}")
        print(f"\tPropreté : {self.cleanliness}\n")

# Temps de test
my_dog = Dog("Vanille")

# Effectuer des activités
my_dog.status()
my_dog.play()  # 1
my_dog.status()
my_dog.eat()   # 2
my_dog.status()
my_dog.sleep() # 3
my_dog.status()
my_dog.walk()  # 4
my_dog.status()
my_dog.take_bath()  # 5
my_dog.status()
my_dog.play()  # 6
my_dog.status()
my_dog.walk()  # 7
my_dog.status()
my_dog.sleep() # 8
my_dog.status()
my_dog.eat()   # 9
my_dog.status()
my_dog.take_bath()  # 10
my_dog.status()
```
